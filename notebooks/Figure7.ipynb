# Figure 7

Example distribution of predictions $s(t)$ of the active predictors in an arbitrary trial, for $k$ in the set of values $\{1,3 ,6,9,12\}$.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from tqdm.auto import tqdm


import warnings
warnings.filterwarnings('ignore')

import sys
sys.path.append('../el_farol')

from main import main
from sim_utils import InteractiveBar
from data_utils import IQR, Commonality
from config import PATHS
```

```python
def Figure7(seed=42, to_file=False):
    ks = [1,3,6,9,12]
    df = main(
        ks=ks,
        ds=[24],
        num_rounds=100,
        num_experiments=1,
        seed=seed
    )

    fig, ax = plt.subplots(
        1,len(ks), 
        figsize=(3*len(ks), 3), 
        tight_layout=True,
        sharey=True,
    )

    for i, k in enumerate(ks):
        df2C = df[df['Num_predictors'] == k].reset_index()
        sns.histplot(
            df2C['Prediction'], bins=30,
            ax=ax[i], kde=True,
        )
        ax[i].set_ylabel('Prediction')
        ax[i].set_xlabel('Round')
        ax[i].set_title(f"k={k}")
        ax[i].grid()
        ax[i].axvline(x=60, ls='--', color='red', linewidth=2)

    if to_file:
        file_name = PATHS['folder_figures_for_paper'] / 'Figure7.png'
        fig.savefig(file_name, dpi=300, bbox_inches='tight')
```

```python
Figure7(to_file=True)
```
