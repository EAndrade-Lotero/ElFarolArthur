# Figure 6

Mean prediction of active predictors vs.~round in a typical simulation for different values of $k\,{=}\,1,3,12$.}

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

import sys
sys.path.append('../el_farol')

from main import main
from config import PATHS
```

```python
def Figure6(seed=42, to_file=False):
    ks = [1,3,12]
    df = main(
        ks=ks,
        ds=[24],
        num_rounds=100,
        num_experiments=1,
        seed=seed
    )

    df2B = df.groupby(['Identifier', 'Round', 'Num_predictors', 'Memory']).agg(
        Prediction=pd.NamedAgg(column="Prediction", aggfunc=lambda x: np.mean(x.values))
    ).reset_index()

    fig, ax = plt.subplots(
        1,3, 
        figsize=(9, 3), 
        tight_layout=True,
    )

    for i, k in enumerate(ks):
        df2C = df2B[df2B['Num_predictors'] == k].reset_index()
        sns.lineplot(
            x='Round', y='Prediction',
            data=df2C,
            ax=ax[i],
            err_style="bars",ci=95
        )
        ax[i].set_ylabel('Prediction')
        ax[i].set_xlabel('Round')
        ax[i].set_title(f"k={k}")
        ax[i].grid()
        ax[i].set_ylim(-5, 105)
        ax[i].axhline(60, ls='--', color='red', linewidth=2)
    if to_file:
        file_name = PATHS['folder_figures_for_paper'] / 'Figure6.png'
        fig.savefig(file_name, dpi=300, bbox_inches='tight')
```

```python
Figure6(to_file=True)
```
