```python
import Procedatos as P
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from heapdict import heapdict

def diferencia(x, y):
    if len(x) == 0:
        return np.nan
    elif len(x) < 2:
        return 0
    else:
        y1 = heapdict()
        for i, j in enumerate(y):
            y1[i] = 100 - j
        ind_max1, i = y1.popitem()
        ind_max2, i = y1.popitem()
#        return f'|{x[ind_max1]} - {x[ind_max2]}|'
        return np.abs(x[ind_max1] - x[ind_max2])
```

```python
data = P.leer_datos([12],[1],[100],[100],
                        verb=False,muchos=False,cola=False)

ids = data['Identificador'].unique()
ident = ids[0]
df = pd.DataFrame(data[data['Identificador'] == ident])

df['Prediccion_lag'] = df.groupby('Agente')['Prediccion'].transform('shift',1)
df1 = df.groupby('Ronda')['Prediccion_lag'].value_counts().reset_index(name='Conteo')

df2 = df1.groupby('Ronda').apply(
#    lambda x: np.sum(x['Prediccion_lag'])
    lambda x: diferencia(list(x['Prediccion_lag']), list(x['Conteo'])) #diferencia(x['Prediccion_lag'], x['Conteo'])
).reset_index()
df2.columns = ['Ronda','Dif_modes']
sns.lineplot(x='Ronda',y='Dif_modes',data=df2)
```
