# Paradoja de la precisión

Al aumentar la cantidad de predictores en la bolsa de cada agente, las predicciones en cada ronda se harán con predictores más precisos. El resultado, paradójicamente, es que la precisión disminuye.

```python
data1 = P.leer_datos([12],[1],[100],[100],
                    verb=False,muchos=False,cola=False)
data1['Modelo'] = '12-1'
data2 = P.leer_datos([12],[12],[100],[100],
                    verb=False,muchos=False,cola=False)
data2['Modelo'] = '12-12'
```

```python
df1 = data1.groupby(['Identificador','Ronda'])['Precision'].mean().reset_index()
df2 = data2.groupby(['Identificador','Ronda'])['Precision'].mean().reset_index()
fig, ax = plt.subplots(1,2, figsize=(6,3), tight_layout=True)
sns.lineplot(x='Ronda',y='Precision',hue='Identificador',data=df1,ax=ax[0])
sns.lineplot(x='Ronda',y='Precision',hue='Identificador',data=df2,ax=ax[1])
ax[0].set_ylim([5,65])
ax[1].set_ylim([5,65])
ax[0].set_ylabel('Inaccuracy')
ax[1].set_ylabel('Inaccuracy')
ax[0].get_legend().remove()
ax[1].get_legend().remove()
ax[0].set_title('12-1')
ax[1].set_title('12-12')
```

Primero, observe que en las rondas iniciales, y sin importar el número de predictores, la precisión mejora.

Segundo, el Inaccuracy del modelo en el cual cada agente tiene solo un predictor oscila menos y es menor respecto al modelo en el cual cada agente tiene 12 predictores de dónde escoger.

Investigaremos primero por qué aumentar el número de predictores en la bolsa aumenta el Deviation. Luego usaremos este resultado para explicar el aumento del Inaccuracy.

# Hipótesis aumento del Deviation

La hipótesis para explicar el aumento de la Deviation es la siguiente:

1. Aumentar el número de predictores tiene como consecuencia que los agentes toman decisiones más parecidas.

Si (1) es cierta, entonces la explicación es clara. Decisiones más parecidas implican mayor sincronía en el comportamiento de los agentes, lo cual aumenta el Deviation.

Vamos a verificar en las simulaciones si hay evidencia de (1). Observemos las predicciones típicas en una ronda para cada uno de los dos modelos en cuestión:

```python
ide = data1.Identificador.unique()[30]
ronda = 95
df1 = data1[(data1.Identificador == ide) & (data1.Ronda == ronda)]
print('Kurtosis 1 predictor:', kurtosis(df1.Prediccion))
ide = data2.Identificador.unique()[20]
ronda = 82
df2 = data2[(data2.Identificador == ide) & (data2.Ronda == ronda)]
print('Kurtosis 12 predictores:', kurtosis(df2.Prediccion))
fig, ax = plt.subplots(1,2, figsize=(6,3), tight_layout=True)
sns.histplot(df1.Prediccion,ax=ax[0],kde=True)
sns.histplot(df2.Prediccion,ax=ax[1],kde=True)
ax[0].set_title('12-1')
ax[1].set_title('12-12')
fig.suptitle('Distribuciones típicas de Prediccion')
```

Con un solo predictor, las predicciones suelen dividirse en dos agrupamientos, mostrando distribuciones bimodales (figura de la izquierda). Por otro lado, con 12 predictores, las predicciones se aglutinan en un solo valor, mostrando distribuciones unimodales (figura de la derecha).

Una manera de medir este comportamiento es mediante la kurtosis de la distribución de las predicciones en cada ronda. La kurtosis mide qué tan larga son las colas de la distribución. En una distribución que concentra más los valores, la kurtosis será mayor. Así pues, si aumentar el número de predictores concentra las predicciones hacia un valor, entonces las kurtosis por ronda para este modelo serán mayores. Esto es lo que, en efecto, se observa. En la siguiente figura mostramos la frecuencia de las kurtosis por ronda para cada uno de los modelos:

```python
df1 = data1.groupby(['Identificador','Ronda'])['Prediccion'].apply(lambda x: kurtosis(x)).reset_index(name='Kurtosis')
df1['Model'] = '12-1'
df2 = data2.groupby(['Identificador','Ronda'])['Prediccion'].apply(lambda x: kurtosis(x)).reset_index(name='Kurtosis')
df2['Model'] = '12-12'
df = pd.concat([df1,df2])
fig, ax = plt.subplots()
sns.histplot(x='Kurtosis',hue='Model',data=df, kde=True, ax=ax)
ax.set_xlim([-2,5])
fig.suptitle('Frecuencias de las\n kurtosis por ronda')
```

¿Por qué las predicciones tienden a un mismo valor a medida que aumenta el número de predictores en la bolsa? Creemos que hay tres razones para este comportamiento.

La primera razón está basada en la siguiente observación. Se tiene que predictores que tienden a dar valores cercanos a 60 tienden a ser más precisos (toda vez que el Inacurracy es mayor cuando la asistencia es 90 para un predictor que predice 20 que para uno que predice 60; o cuando la asistencia es de 20 para un predictor que predice 90 que para el que predice 60). Por lo tanto, predictores más precisos tienden a dar valores más cercanos a 60, como se observa en la siguiente figura:

```python
dfA = data1.groupby(['Identificador','Politica','Ronda']).agg({'Prediccion':'mean', 'Precision':'mean'}).reset_index()
dfA['Inaccuracy'] = dfA['Precision']
del dfA['Precision']
dfB = data2.groupby(['Identificador','Politica','Ronda']).agg({'Prediccion':'mean', 'Precision':'mean'}).reset_index()
dfB['Inaccuracy'] = dfB['Precision']
del dfB['Precision']
fig, ax = plt.subplots(1,2,figsize=(6,3),tight_layout=True)
sns.scatterplot(x='Inaccuracy',y='Prediccion',alpha=.25,marker='+',
                data=dfA,ax=ax[0])
sns.scatterplot(x='Inaccuracy',y='Prediccion',alpha=.25,marker='+',
                data=dfB,ax=ax[1])
ax[0].set_title('12-1')
ax[1].set_title('12-12')
```

En esta figura, cada datapoint es una política en una ronda en un trial. Los gráficos muestran el valor de la predicción que dió la política en dicha ronda y la precisión que tenía en dicha ronda. En ambos modelos se observa una tendencia según la cual las políticas más precisas dan una predicción cercana a 60.

De esta manera, cuanto más ámplia la bolsa de predictores, los agentes predicen con base en predictores más precisos. Entonces muchas predicciones se aglutinarán en torno a valores cercanos a 60.

Lo segundo es que la cantidad de predictores activos disminuye a medida que aumenta la bolsa de predictores, como podemos ver en la siguiente figura:

```python
data = P.leer_datos([12],[1,3,6,9,12],[100],[100],
                    verb=False,muchos=False,cola=True)
df = data.groupby(['Num_predic','Identificador'])['Politica'].nunique().reset_index(name='Num_pred_act')
sns.lineplot(x='Num_predic',y='Num_pred_act',data=df)
```

Esto se debe a que hay un número limitado de predictores. Cuanto más aumenta el tamaño de la bolsa de cada agente, más predictores van a compartir. Varios de esos predictores resultarán muy precisos, por lo que más agentes tendrán los mismos predictores. Las predicciones que los agentes harán con base en ellos serán las mismas.

```python
import ElFarol_Arthur as E

num_agentes = 100
num_rondas = 100
umbral = .6
num_predictores = 12
long_memoria = 12
bar2 = E.simulacion(num_agentes,umbral,long_memoria,num_predictores,
                    num_rondas,DEB=False,to_file=False)
```

```python
ronda = 45
agentes = bar2.agentes
predictores = [a.predictores for a in agentes]
predictores = [p for sublist in predictores for p in sublist]
predictores = list(set(predictores))
rotulos_predictores = [str(p) for p in predictores]
predictores_activos1 = [a.predictor_activo[ronda] for a in agentes]
predictores_activos = list(set(predictores_activos1))
rotulos_predictores_activos = [str(p) for p in predictores_activos]
inaccs_todos = np.mean([p.precision[ronda+1] for p in predictores])
inaccs_activos = np.mean([p.precision[ronda+1] for p in predictores_activos])
predicciones = [p.prediccion[ronda] for p in predictores]
precisiones = [p.precision[ronda] for p in predictores]
print('# predictores:', len(predictores))
print('# predictores activos:', len(predictores_activos))
print('Media de inaccuracy todos los predictores:', inaccs_todos)
print('Media de inaccuracy predictores activos:', inaccs_activos)
data = pd.DataFrame({'Predictor':rotulos_predictores,
                     'Prediccion':predicciones,
                     'Inaccuracy':precisiones})
data['Activo'] = data['Predictor'].apply(lambda x: 'Sí' if x in rotulos_predictores_activos else 'No')
fig, ax = plt.subplots()
sns.scatterplot(x='Inaccuracy', y='Prediccion', hue='Activo',
               alpha=.8, marker='+', data=data)
ax.set_xlim([20,50])
ax.set_ylim([0,100])
```

Una tercera razón es que no es solo un efecto causado por usar predictores muy precisios, sino que tiene que ver con la definición misma de precisión. Dos políticas que tengan predicciones parecidas en el transcurso de las rondas, tendrán precisiones parecidas. Claro, están aquellos pares de predictores que dan predicciones en espejo y ellas también tendrán precisiones parecidas. Pero el hecho es que hay una buena probabilidad de que dos predictores con precisiones parecidas den predicciones parecidas.
