```python
import pandas as pd
import Procedatos as P
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = P.leer_datos([12],[12],[100],[100],
                        verb=False,muchos=False,cola=False)
```

```python
data.head()
```

```python
ids = data['Identificador'].unique()

ident = ids[0]

for ronda in range(81, 100):
    fig, ax = plt.subplots(1,4, figsize=(12,3), tight_layout=True)

    df = pd.DataFrame(data[data['Ronda'] == ronda])
    df['Asistencia'] = df.groupby('Identificador')['Estado'].transform('sum')
    df = pd.DataFrame(df[df['Identificador'] == ident])
    df['P-A'] = df['Prediccion'] - df['Asistencia']

    sns.histplot(x='Prediccion', data=df, kde=True, ax=ax[0])
    sns.histplot(x='Estado', data=df, kde=True, ax=ax[1])
    sns.histplot(x='Puntaje', data=df, kde=True, ax=ax[2])
    sns.histplot(x='P-A', data=df, kde=True, ax=ax[3])
    
    fig.suptitle(f'Ronda: {ronda}')
```
